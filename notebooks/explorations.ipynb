# Senoide

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 1, 0.01)     # tempo
f = 5                         # frequência de 5 Hz
y = np.sin(2 * np.pi * f * t) # fórmula da onda senoidal

plt.plot(t, y)
plt.title("Onda senoidal de 5 Hz")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

print(t)
```

# Onda

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 1, 0.001)   # tempo (1 segundo com 1000 pontos)
f1 = 5                       # frequência da primeira onda
f2 = 7                       # frequência da segunda onda

y1 = np.sin(2 * np.pi * f1 * t)
y2 = np.sin(2 * np.pi * f2 * t)
y_soma = y1 + y2             # soma das duas ondas

plt.figure(figsize=(10, 6))
plt.plot(t, y1, label="Onda 1 (5 Hz)")
plt.plot(t, y2, label="Onda 2 (7 Hz)")
plt.plot(t, y_soma, label="Soma (interferência)", linewidth=2)
plt.legend()
plt.title("Soma de Duas Ondas Senoidais")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# parâmetros
t = np.arange(0, 1, 0.001)  # 1000 amostras (1 segundo)
f1 = 5                      # primeira frequência
f2 = 20                     # segunda frequência

# soma das duas ondas
y = np.sin(2 * np.pi * f1 * t) + np.sin(2 * np.pi * f2 * t)

plt.figure(figsize=(10, 4))
plt.plot(t, y)
plt.title("Onda composta (5 Hz + 20 Hz)")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

# FFT
n = len(y)               # número de amostras
freqs = np.fft.fftfreq(n, d=0.001)  # eixo de frequência (Hz)
fft = np.fft.fft(y)      # transforma o sinal

# módulo da FFT (magnitude)
magnitude = np.abs(fft)

plt.figure(figsize=(10, 4))
plt.plot(freqs[:n//2], magnitude[:n//2])  # só a metade positiva
plt.title("Espectro de Frequências")
plt.xlabel("Frequência (Hz)")
plt.ylabel("Intensidade")
plt.grid(True)
plt.show()
```

# Amostragem

```python
import numpy as np
import matplotlib.pyplot as plt

f = 5                    # frequência da onda (Hz)
t_fina = np.arange(0, 1, 0.001)   # amostragem alta (1000 Hz)
t_grossa = np.arange(0, 1, 0.05)  # amostragem baixa (20 Hz)

y_real = np.sin(2 * np.pi * f * t_fina)
y_amostrada = np.sin(2 * np.pi * f * t_grossa)

plt.figure(figsize=(10, 5))
plt.plot(t_fina, y_real, label="Onda real (contínua)")
plt.scatter(t_grossa, y_amostrada, color='red', label="Amostras (poucas)")
plt.legend()
plt.title("Amostragem: contínua vs discreta")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```
